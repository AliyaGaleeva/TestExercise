```python
import pandas as pd
import numpy as np
import string
```

```python
df = pd.read_excel('Prilozhenie.xlsx', 'DIM_IDV_LBY_AR')
df
```

```python
str1 = "INSERT DIM_IDV_LBY_AR(DIM_EFF_DT, DIM_END_DT, LBY_AR_ID, CST_ID, PD_ID, PRC)\nVALUES\n"
with open('fill_table_DIM_IDV_LBY_AR.sql', 'a', encoding="utf-8") as filehandle:
        filehandle.write('%s\n' % str1)
for i in range (15):
    str1 = "('"+ str(df.iloc[i].DIM_EFF_DT) + "', '"  + str(df.iloc[i].DIM_END_DT) + "', "  + str(df.iloc[i].LBY_AR_ID) + ", " + str(df.iloc[i].CST_ID) + ", " + str(df.iloc[i].PD_ID) + ", " + np.format_float_positional(df.iloc[i].PRC) + ")"
    if i==14:
        str1 = str1 + ";"
    else:
        str1 = str1 + ","
    with open('fill_table_DIM_IDV_LBY_AR.sql', 'a', encoding="utf-8") as filehandle:
        filehandle.write('%s\n' % str1)
```
