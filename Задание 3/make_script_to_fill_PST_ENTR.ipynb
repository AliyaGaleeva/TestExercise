```python
for k in range(70):
    file = "fill_table_PST_ENTR" + str(k) + ".sql"
    str1 = "INSERT PST_ENTR(PST_ENTR_ID, PST_DT, DB_AC_NBR, CR_AC_NBR, DB_AMT)\nVALUES\n"
    with open(file, 'a', encoding="utf-8") as filehandle:
        filehandle.write('%s\n' % str1)
    for i in range (300*k, 300*(k+1)):
        str1 = "("+ str(df.iloc[i].PST_ENTR_ID) + ", '" + str(df.iloc[i].PST_DT) + "', '"  + str(df.iloc[i].DB_AC_NBR) + "', '" + str(df.iloc[i].CR_AC_NBR) + "', "  + str(df.iloc[i].DB_AMT) + ")"
        if i==300*(k+1)-1:
            str1 = str1 + ";"
        else:
            str1 = str1 + ","
        with open(file, 'a', encoding="utf-8") as filehandle:
            filehandle.write('%s\n' % str1)
```
